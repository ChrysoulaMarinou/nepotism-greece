```python
#imports
from bs4 import BeautifulSoup
import requests
import time

import pandas as pd

#set ups
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
response = requests.get('https://www.hellenicparliament.gr/Vouli-ton-Ellinon/To-Politevma/Ekloges/Eklogika-apotelesmata-New/')
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
urls = []

paragraphs = soup.find_all('p')

for paragraph in paragraphs:
    links = paragraph.find_all('a',{'target':'_blank'})
    
    for link in links:
        if 'Σύνθεση' in link.text:  
            url= link.get('href')
            print(url)
            urls.append(url)  
```

```python
more_urls = []

for p in soup.find_all('p'):
    if 'Σύνθεση' in p.text:
        url = p.find('a',href=True).get('href')
        print(url)

        more_urls.append(url)
        
```

```python
len(more_urls)
```

```python
temp_df= pd.DataFrame(more_urls[1:])
temp_df= temp_df.rename(columns={temp_df.columns[0]:'url'})
temp_df.head()
```

```python
base_url = 'http://www.hellenicparliament.gr'
```

```python
temp_df['final_url']=''

for i,item in temp_df.iterrows():
    if 'hellenicparliament.gr/' in item['url']:
        temp_df.at[i, 'final_url'] = item['url']
        
    else:
        temp_df.at[i, 'final_url'] = base_url+item['url']
```

```python
temp_df
```

```python
temp_df.final_url[0]
```

```python
numbers = list(range(1,31))
```

```python
len(numbers)
```

```python
numbers[-1]
```

```python
webpages=[]

for url in temp_df.final_url[0:17]:
    if 'SInthesi-Proigoumenon-Periodon' in url:
        for n in numbers[0:15]:
            webpage = url + '&pageNo=' + str(n)
            webpages.append(webpage)
            
    else:
        for n in numbers[0:30]:
            webpage = url+'?pageNo='+str(n)

            webpages.append(webpage)
            
webpages
```

```python
len(webpages)
```

```python
df1=pd.DataFrame(webpages)
df1=df1.rename(columns={df1.columns[0]:'url'})
df1
```

```python
df1.url[140]
```

```python
#Οι κοινοβουλευτικές περιόδοι έχουν αντικαταστάσεις βουλευτών οι οποίες παρατίθονται 
#σε ξεχωριστό πίνακα με ξεχωριστο url που ορίζεται με &pageNoR=
webpages2=[]

for url in df1.url:
    if '&pageNo=15' in url:
        for n in numbers[0:2]:
            webpage = url + '&pageNoR=' + str(n)
            webpages2.append(webpage)
            
    
webpages2
```

```python
#df2 περιλαμβάνει τα ονόματα των αντικαταστατων βουλευτών
df2=pd.DataFrame(webpages2)
df2=df2.rename(columns={df2.columns[0]:'url2'})
df2
```

```python
names = []

#loop through your urls
#for each one of those
for url in df1.url:
    #make a request and visit it
    response = requests.get(url)
    #make python your script sleep for 3 seconds before making the next request
    time.sleep(3)
    #parse each one of the pages
    soup_doc = BeautifulSoup(response.text, 'html.parser')
    
#     try:
    #grab all the rows found in the table on each one of the webpages you visit
    rows = soup_doc.find('table').find('tbody').find_all('tr')
#     except:
#         pass
    
    #grap all the cell values for each one of the rows you got but for the last row in each table (with is the pagination bar)
    for row in rows[:-1]:
        cells  = row.find_all('td')

        if 'SInthesi-Proigoumenon-Periodon' in url:
            for cell in cells[0::3]:
                print(cell.text.strip())
                names.append(cell.text.strip())
                
        else:
#           loop through your cells that hold MP's names
            for cell in cells[1::3]:
#               print the names as you scrape them
                print(cell.text)
#               append your names list with the cell values (names) you are getting
                names.append(cell.text.strip())
```

```python
df_names=pd.DataFrame(names)

try:
    df_names=df_names.rename(columns={df_names.columns[0]:'names'})
    df_names=df_names[['Last Name','First Name','Father Name',]] = df_names.names.str.split(' ', expand = True)
    
except:
    pass

#df_names[['first_name', 'last_name','father_name']] = df_names.apply(lambda row: row['names'].split(), axis=1)
df_names
```
