```python
#imports
from bs4 import BeautifulSoup
import requests
import time

import pandas as pd

#set ups
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
response = requests.get('https://www.hellenicparliament.gr/Vouli-ton-Ellinon/To-Politevma/Ekloges/Eklogika-apotelesmata-New/')
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
urls = []

paragraphs = soup.find_all('p')

for paragraph in paragraphs:
    links = paragraph.find_all('a',{'target':'_blank'})
    
    for link in links:
        if 'Σύνθεση' in link.text:  
            url= link.get('href')
            print(url)
            urls.append(url)  
```

```python
more_urls = []

for p in soup.find_all('p'):
    if 'Σύνθεση' in p.text:
        url = p.find('a',href=True).get('href')
        print(url)

        more_urls.append(url)
        
```

```python
len(urls)
```

```python
temp_df= pd.DataFrame(more_urls[1:])
temp_df= temp_df.rename(columns={temp_df.columns[0]:'url'})
temp_df.head()
```

```python
len(temp_df)
```

```python
base_url = 'http://www.hellenicparliament.gr'
```

```python
#  φτιάχνουμε το νέο url, προσθέτωντας το 'http://www.hellenicparliament.gr' σε νέο column στο df
temp_df['final_url']=''

for i, item in temp_df.iterrows():
    if 'hellenicparliament.gr/' in item['url']:
        temp_df.at[i, 'final_url'] = item['url']
        
    else:
        temp_df.at[i, 'final_url'] = base_url + item['url']
```

```python
temp_df
```

```python
numbers = list(range(1,31))
```

```python
len(numbers)
```

```python
numbers[-1]
```

```python
temp_df.final_url[0]
```

```python
webpages=[]

# στην στήλη που έχουμε φτιάξει τα νέα url 
for url in temp_df.final_url[0:18]:

    if 'SInthesi-Proigoumenon-Periodon' in url:
        
#         πρόσθεσε στο url που αντίστοιχεί στη θέση που βρήκες το &pageNo= και το νούμερο
        for n in numbers[0:15]:
            webpage = url + '&pageNo=' + str(n)
            webpages.append(webpage)
            
    else:
#         αλλιώς βάλε τον άλλο σχηματισμό ?pageNo=
        for n in numbers[0:30]:
            webpage = url + '?pageNo=' + str(n)

            webpages.append(webpage)
            
webpages
```

```python
len(webpages)
```

```python
# αποθήκευσέ τα σε ένα df
df1=pd.DataFrame(webpages)
df1=df1.rename(columns={df1.columns[0]:'url'})
df1
```

```python
df1.url[140]
```

```python
#Οι κοινοβουλευτικές περιόδοι έχουν αντικαταστάσεις βουλευτών οι οποίες παρατίθονται 
#σε ξεχωριστό πίνακα με ξεχωριστο url που ορίζεται με &pageNoR=
webpages2=[]

for url in df1.url:
    if '&pageNo=15' in url:
        for n in numbers[0:2]:
            webpage = url + '&pageNoR=' + str(n)
            webpages2.append(webpage)
            
    
webpages2
```

```python
#df2 περιλαμβάνει τα ονόματα των αντικαταστατων βουλευτών
df2=pd.DataFrame(webpages2)
df2=df2.rename(columns={df2.columns[0]:'url'})
df2
```

```python
# κάνε τα ένα df
# αυτό σημαίνει πως δεν έχεις ξεχωρίσει ποιοί εκλέγονται από τον λαό και ποιοί επιλέγονται από το κόμμα
all_urls = pd.concat([df1, df2], axis = 0,  ignore_index=True)
```

```python
names = []

#loop through your urls
#for each one of those
for url in all_urls.url:
    #make a request and visit it
    response = requests.get(url)
    #make python your script sleep for 3 seconds before making the next request
    time.sleep(3)
    #parse each one of the pages
    soup_doc = BeautifulSoup(response.text, 'html.parser')
    
#     try:
    #grab all the rows found in the table on each one of the webpages you visit
    rows = soup_doc.find('table').find('tbody').find_all('tr')
#     except:
#         pass
    
    #grap all the cell values for each one of the rows you got but for the last row in each table (with is the pagination bar)
    for row in rows[:-1]:
        cells  = row.find_all('td')

        if 'SInthesi-Proigoumenon-Periodon' in url:
            for cell in cells[0::3]:
                print(cell.text.strip())
                names.append(cell.text.strip())
                
        else:
#           loop through your cells that hold MP's names
            for cell in cells[1::3]:
#               print the names as you scrape them
                print(cell.text)
#               append your names list with the cell values (names) you are getting
                names.append(cell.text.strip())
```

```python
df_names=pd.DataFrame(names)
```

```python
len(df_names)
```

```python
# αποθήκευσα τη λίστα με τα ονόματα ώστε να μην scrapάρω διαρκώς
# df_names.to_csv('mps2024.csv',encoding='utf-8',index=False)
df_names_one_occurance = df_names.drop_duplicates(keep='first').reset_index(drop=True).copy()
```

```python
# επανονοματίζω την στήλη 0
df_names_one_occurance = df_names_one_occurance.rename(columns={df_names_one_occurance.columns[0]:'names'}).copy()
```

```python
# άρα τις περισσότερες φορές στο κοινοβούλιο εκλέγονται τα ίδια άτομα
len(df_names_one_occurance)
```

```python
# ξεφωρτόνομαι το double space και την παύλα (-)
df_names_one_occurance['names'] = df_names_one_occurance.names.str.replace('-',' ').str.strip()
df_names_one_occurance['names'] = df_names_one_occurance.names.str.replace('  ','').str.strip()
```

```python
df_names_one_occurance.to_csv('mps2024_cleaned.csv',encoding='utf-8',index=False)
```

```python
# # ΔΙΑΒΑΣΜΑ
# try:
#     df_names[['Last Name','First Name','Father Name',]] = df_names.names.str.split(' ', expand = True)
    
# except:
#     pass

df_names_one_occurance[['first_name', 'last_name','father_name']] = df_names.apply(lambda row: row['names'].split(), axis=1)
df_names_one_occurance
```

```python
# Function to split the names and ensure exactly three parts
def split_names(row):
    parts = row['names'].split()
    # Ensure exactly three parts, fill with None if needed
    if len(parts) < 4:
        parts += [None] * (4 - len(parts))
    return pd.Series(parts, index=['first_name', 'last_name', 'father_name','else'])

# Apply the function to split the names into three columns
df_names_one_occurance[['first_name', 'last_name', 'father_name']] = df_names_one_occurance.apply(split_names, axis=1)

# Display the resulting DataFrame
print(df_names_one_occurance)
```
