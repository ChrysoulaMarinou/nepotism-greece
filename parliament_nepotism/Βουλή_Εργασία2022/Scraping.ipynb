```python
url = 'https://covid19.innews.gr/iframe'
data = get(url)
content = BeautifulSoup(data.text, "html.parser")
```

```python
contents = [item for item in content.select("div.number-container-content")]
```

```python
# create an empty list, in which we'll save a series of dictionaries
results = []
# contents is a list. For every item of this list
for content in contents:
#     create an empty dictionary
    content_dict = {}
#     save in the title variable all the items which have the .title selector ... etc
    title = content.select('.title')
#     chained_title = list(itertools.chain.from_iterable(content.select('.title')))
    
    total = content.select('.total.number')
#     chained_total = list(itertools.chain.from_iterable(content.select('.total.number')))
#     print(chained_total)

    today = content.select('.today.number')
#     chained_today = list(itertools.chain.from_iterable(content.select('.today.number')))
    
#     create a key to the empty dictionary which has the value of the variable title ... etc
    content_dict['title'] = title
    content_dict['total'] = total
    content_dict['today'] = today
#     save each dictionary to the list results
    results.append(content_dict)
```

```python
df = pd.DataFrame(results)
df
```
